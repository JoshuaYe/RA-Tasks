```python
# Descriptive statistics by quarter
quarters = ['q1', 'q2', 'q3', 'q4']
for year in range(2002, 2021):
    for q in quarters:
        date = str(year) + q
        if not firm_risk[firm_risk.date == date].empty:
            print("\nDescriptive statistics for", date)
            print(firm_risk[['PRisk', 'NPRisk', 'Risk', 'PSentiment', 'NPSentiment', 'PRiskT_economic', 'PRiskT_environment', 'PRiskT_trade', 'PRiskT_institutions', 'PRiskT_health', 'PRiskT_security', 'PRiskT_tax', 'PRiskT_technology', 'Covid_Exposure', 'Covid_Neg_Sentiment', 'Covid_Pos_Sentiment', 'Covid_Net_Sentiment', 'Covid_Risk', 'SARS_Exposure', 'H1N1_Exposure', 'Zika_Exposure', 'Ebola_Exposure', 'Brexit_Exposure', 'Brexit_Neg_Sentiment', 'Brexit_Pos_Sentiment', 'Brexit_Net_Sentiment', 'Brexit_Risk']][firm_risk.date == date].describe().round(2))
```
